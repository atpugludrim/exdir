1. Compute $\pi$ as $2V_{k,n}$.

```python
n=1
k=49
n=n+k
```

```python
from decimal import Decimal, getcontext

ctx=getcontext()
ctx.prec=646
```

```python
V = {0:[]}
p = Decimal('1.')
partial_denom=Decimal('0.')
for i in range(n):
    partial_denom = ctx.sqrt(Decimal('2')+partial_denom)
    p *= Decimal('2')/partial_denom
    V[0].append(p)
print(2*p)
```

```python
for i in range(1, k + 1):
    if i not in V:
        V[i] = []
    for j in range(n - 1):
        try:
            Vkn = (ctx.power(Decimal('4'),Decimal(i))*V[i-1][j+1]-V[i-1][j])/(ctx.power(Decimal('4'),Decimal(i))-Decimal('1'))
            V[i].append(Vkn)
        except IndexError:
            break
#             import pdb;pdb.set_trace()
# import pdb;pdb.set_trace()
print(Decimal('2')*V[k][-1])
```
